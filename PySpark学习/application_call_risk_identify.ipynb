````python
# time模块的基础用法
time.time() # 获取当前时间戳
time.ctime() # 获取当前时间戳的字符串格式
time.localtime() # 获取当前时间的结构化对象
time.gmtime() # 获取当前时间的结构化对象
# struct_time 是 time 模块的「结构化时间对象」，包含年、月、日、时、分、秒等字段，便于提取具体时间信息。
# 1. 时间戳 → struct_time（本地时区）
timestamp = time.time()
local_struct = time.localtime(timestamp)
print(local_struct.tm_year)  # 提取年份：2024
print(local_struct.tm_hour)  # 提取小时：14（本地时区）

# 2. struct_time → 时间戳
timestamp2 = time.mktime(local_struct)
print(timestamp2)  # mktime 会忽略微秒

# 3. struct_time → 格式化字符串（用 strftime）
fmt_str = time.strftime("%Y-%m-%d %H:%M:%S", local_struct)
print(fmt_str)  # 2025-10-28 19:39:32

# 4. 字符串 → struct_time（用 strptime，需指定格式）
struct_from_str = time.strptime(fmt_str, "%Y-%m-%d %H:%M:%S")
print(struct_from_str.tm_mday)  
````

````python
from datetime import datetime, date, time
# datetime模块基础用法
# 1. 获取当前「本地日期+时间」
now_local = datetime.now()
print(now_local)  

# 2. 获取当前「UTC 日期+时间」（避免时区偏差）
now_utc = datetime.utcnow()
print(now_utc) 

# 3. 手动创建 datetime 对象（年, 月, 日, 时, 分, 秒, 微秒）
dt = datetime(2024, 6, 4, 14, 41, 18, 123456)
print(dt)  # 2024-06-04 14:41:18.123456

# 4. 单独创建 date / time 对象
d = date(2024, 6, 4)
t = time(14, 41, 18)
# 合并为 datetime 对象
dt2 = datetime.combine(d, t)
print(dt2)  # 2024-06-04 14:41:00


dt = datetime(2024, 6, 4, 14, 41, 18, 123456)
# 1. datetime → 字符串（strftime）
fmt1 = dt.strftime("%Y-%m-%d %H:%M:%S")
print(fmt1)  # 2024-06-04 14:41:18

fmt2 = dt.strftime("%m/%d/%Y %I:%M %p")  # 12小时制+AM/PM
print(fmt2)  # 06/04/2024 02:41 PM

# 2. 字符串 → datetime（strptime，必须严格匹配格式）
str_dt = "2024-06-04 14:41:18"
dt2 = datetime.strptime(str_dt, "%Y-%m-%d %H:%M:%S")
print(dt2.year)  # 2024
print(dt2.hour)  # 14
````

````python
# 时间计算
from datetime import datetime, timedelta

now = datetime.now()

# 1. 计算未来时间（datetime + timedelta）
future_1day = now + timedelta(days=1)  # 1天后
print(future_1day)  

future_3h30m = now + timedelta(hours=3, minutes=30)  # 3小时30分钟后
print(future_3h30m)  

# 2. 计算过去时间（datetime - timedelta）
past_2days = now - timedelta(days=2)  # 2天前
print(past_2days)  

# 3. 计算两个 datetime 的差值（自动返回 timedelta）
dt1 = datetime(2024, 6, 4)
dt2 = datetime(2024, 6, 1)
diff = dt1 - dt2
print(diff)  # 3 days, 0:00:00（timedelta 对象）
print(diff.days)  # 提取天数：3
print(diff.total_seconds())  # 转换为总秒数：259200.0
````

````python
from pyspark.sql.functions import col

# 从指定表中获取当天 etl_dt 且任务名为 "callIdentify" 的最新 task_id
latest_task_id = (
    spark.table("mart_caterb2b_mall_risk_management.friday_offline_batch_infer_record")
         .filter((col("dt") == etl_dt) & (col("task_name") == "callIdentify"))  # 过滤出当天且任务名为 callIdentify 的记录
         .orderBy(col("task_id").desc())   # 按 task_id 降序排列，获取最新记录
         .limit(1)                         # 限制结果只取第一条（即最新的一条）
         .select("task_id")                # 只选择 task_id 字段
         .collect()[0]["task_id"]          # 收集结果并提取 task_id 的值
)

# 检查是否成功获取到 task_id，若无则抛出异常
if not latest_task_id:
    raise RuntimeError(f"当天 {etl_dt} 无 task_id 记录，请先写入数据！")
task_id = latest_task_id
print(task_id)
````

````python
import requests
import json

def query_friday_task(task_id, tenant_id: str = "1666025067693609053"):
    """
    查询 Friday 任务详情
    :param task_id:   任务ID
    :param tenant_id: 租户ID（默认1666025067693609053）
    :return:          接口返回的 JSON（dict）
    """
    url = "https://friday.sankuai.com/api/friday/ml/v1/experiment/eval/api/task/detail"
    params = {"tenantId": tenant_id,"taskId": task_id}
    headers = {"Content-Type": "application/json"}

    resp = requests.get(url,params=params)
    resp.raise_for_status()
    if resp.json()["code"] != 0 or resp.json()["data"] is None:
        raise RuntimeError(f"task_id {task_id} 调用查询任务进度接口异常")
    return resp.json()["data"]
# 示例：查询刚才写入的 task
detail = query_friday_task(task_id)
print(json.dumps(detail, ensure_ascii=False, indent=2))
````

````python
# ---------- 阻塞等待终态 ----------
while True:
    status = detail["status"] # 排队中/运行中/运行失败/运行成功/停止运行
    if status in {"运行失败", "停止运行"}:
        raise RuntimeError(f"task_id {task_id} 运行异常，请重试！")
    if status == "运行成功":
        break
    print("任务运行中或排队中，请等待...")
    time.sleep(5)                              # 每 5s 轮询一次

# ---------- 只有成功才会走到这里 ----------
from pyspark.sql.functions import (
    col, from_json, expr, regexp_replace, when
)
from pyspark.sql.types import StructType, StructField, StringType, ArrayType

df = spark.read.json(detail["output"])

# ---------- 1. 手写 schema（与目标 JSON 一致） ----------
risk_schema = StructType([
    StructField("overlapping_customer_development_risk_result", StringType()),
    StructField("overlapping_customer_development_score",   StringType()),
    StructField("overlapping_customer_development_reason",      StringType()),
    StructField("overlapping_customer_development_evidence",    StringType()),
    StructField("fake_order_risk_result", StringType()),
    StructField("fake_order_score", StringType()),
    StructField("fake_order_reason", StringType()),
    StructField("fake_order_evidence", StringType()),
    StructField("order_proxy_risk_result", StringType()),
    StructField("order_proxy_score", StringType()),
    StructField("order_proxy_reason", StringType()),
    StructField("order_proxy_evidence", StringType())
])

# ---------- 2. 把 Markdown 代码块剥掉，只剩 JSON 串 ----------
# 去掉 ```json 与末尾 ```，同时干掉转义引号
json_str = regexp_replace(
    regexp_replace(col('output.content'), '(?s)```json\\s*', ''),  # 去掉开头
    '(?s)\\s*```', ''                                                # 去掉结尾
)

# ---------- 3. 解析 JSON（失败整行 NULL） ----------
risk_parsed = from_json(
    when(json_str == '', None).otherwise(json_str),   # 空串变 NULL
    risk_schema
)

# ---------- 4. 取 user 消息（同之前，兼容空数组） ----------
user_content_expr = expr("""
    CASE
        WHEN source.messages IS NULL THEN NULL
        WHEN size(filter(source.messages, x -> x.role = 'user')) = 0 THEN NULL
        ELSE filter(source.messages, x -> x.role = 'user')[0].content
    END
""")

# ---------- 5. 组装最终字段 ----------
out_df = df.select(
    col('target'),
    user_content_expr.alias('process_voice'),
    risk_parsed.alias('risk')
).select('target', 'process_voice', 'risk.*')

# ---------- 6. 落表 & 验数 ----------
out_df.createOrReplaceTempView('risk_df')
spark.sql('SELECT target, process_voice, overlapping_customer_development_risk_result, fake_order_risk_result, order_proxy_risk_result FROM risk_df LIMIT 10').show(truncate=False)
````
